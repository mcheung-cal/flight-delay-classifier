# Turn Dependent Variable Depature Delay into Binary Variable

```python
df['DEP_DELAY'] = df['DEP_DELAY'].apply(lambda x: 1 if x > 0 else 0) # 1 if flight delayed 0 o.w
print('percent of flights delayed', sum(df['DEP_DELAY'])/len(df))
```

```python
#convert Dew Point from object to int
df['Dew Point'] = df['Dew Point'].astype(str).astype(int)
df['Dew Point']
#drop innecessary cols
df_new = df.drop(columns=["TAIL_NUM"])
df_new.head()
```

```python
#one-hot encode at categorical cols
df_new = pd.get_dummies(df_new, drop_first=True)
df_new.head()
```

```python
#training and test data split
from sklearn.model_selection import train_test_split

train, test = train_test_split(df_new, test_size=0.3, random_state=88)
x_train = train.drop(columns=["DEP_DELAY"])
y_train = train['DEP_DELAY']
x_test = test.drop(columns=["DEP_DELAY"])
y_test = test['DEP_DELAY']
x_train.shape, x_test.shape
```

# Baseline Model

```python
# Baseline Model - Always predict no delay (0)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

if sum(df['DEP_DELAY']) >= (len(df_new)/2): # if delayed flights >= half total flights always predicted delay (1)
    y_predBL = pd.Series([1 for x in range(len(y_test))])
else: # always predict no delay (0)
    y_predBL = pd.Series([0 for x in range(len(y_test))])

cm_baseline = confusion_matrix(y_test, y_predBL)
baseline_acc = accuracy_score(y_test, y_predBL)
print ("Confusion Matrix: \n", cm_baseline)
print ("\nAccuracy:", baseline_acc)
```

## VIF Test

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

def VIF (df, columns):
    values = sm.add_constant(df[columns]).values
    num_columns = len(columns) + 1
    vif = [variance_inflation_factor(values, i) for i in range(num_columns)]
    return pd.Series(vif[1:], index =  columns)
```

```python
cols = list(df_new.columns.values)
cols.remove('DEP_DELAY')
highVIF = VIF(x_train, cols)
highVIF
```

```python
i = 0
for j in highVIF:
    if j >= 10:
        #x_train = x_train.drop(highVIF.index[i])
        cols.remove(highVIF.index[i])
    i = i + 1
```

```python
#remove highVIF columns
for i in cols:
    x_train = x_train.drop(columns=[i])
    x_test = x_test.drop(columns=[i])
x_train.head()
```

```python
logReg2 = LogisticRegression(random_state = 88).fit(x_train,y_train)
y_log_pred2 = logReg2.predict(x_test)
logreg_acc2 = accuracy_score(y_test, y_log_pred)
print(f'Logistic Regression Test Accuracy: {logreg_acc:.4f}')
```

# Bootstap

```python
from sklearn import metrics

def perform_bootstrap(test_data, test_label, model, sample=500, random_state=88): 
    n_sample = sample
    output_array=np.zeros([n_sample])
    output_array[:]=np.nan
    print(output_array.shape)
    for bs_iter in range(n_sample):
        bs_index = np.random.choice(test_data.index, len(test_data.index), replace=True)
        bs_data = test_data.loc[bs_index]
        bs_label = test_label.loc[bs_index]
        bs_predicted = model.predict(bs_data)
        accuracy = metrics.accuracy_score(bs_label, bs_predicted)
        output_array[bs_iter] = accuracy
    output_df = pd.DataFrame(output_array)
    return output_df
```

```python
boot_lda = perform_bootstrap(x_test, y_test, lda)
boot_cart = perform_bootstrap(x_test, y_test, dtc_cv)
boot_randomForest = perform_bootstrap(x_test, y_test, rf1)
```

```python
import matplotlib.pyplot as plt

plt.hist(boot_lda, label = 'LDA', edgecolor = 'black')
plt.title('Bootstrapped Model Accuracy')
plt.legend(loc='upper right')
plt.show()

plt.hist(boot_cart, label = 'CART', color = 'orange', edgecolor = 'black')
plt.title('Bootstrapped Model Accuracy')
plt.legend(loc='upper right')
plt.show()

plt.hist(boot_randomForest, label = 'Random Forest', color = 'green', edgecolor = 'black')
plt.title('Bootstrapped Model Accuracy')
plt.legend(loc='upper right')
plt.show()

```
